```python
import pandas as pd
import numpy as np
movie = pd.read_csv("movie_metadata.csv")
movie.info()
```

```python
movie.imdb_score
```

#as one can see, imdb_scores are discrete *variables*

```python
def greater_than_7(imdb_score):
  if imdb_score > 7.0:
    return 1
  else:
    return 2
```

```python
movie['imdb_scores_greater_than_7'] = movie['imdb_score'].apply(greater_than_7)
display(movie.head())
movie.groupby("imdb_scores_greater_than_7")["imdb_score"].count()
```

# Hence there are 1594 movies that have imdb score greater than 7, whereas 3449 movies have less than 7.
