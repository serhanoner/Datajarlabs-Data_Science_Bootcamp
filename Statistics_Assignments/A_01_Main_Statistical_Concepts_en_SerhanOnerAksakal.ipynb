# Mode

```python
def mode(x):
  y={}
  for a in x:
    if not a in y:
      y[a]=1 
    else:
      y[a]+=1 
  return [g for g,l in y.items() if l==max(y.values())]

print(mode(df['age']))
print(mode(df['salary']))
print(mode(df['happiness_rate']))
```

# Variances

```python
print(np.var(df['age'], ddof=1))
print(np.var(df['salary'], ddof=1))
print(np.var(df['happiness_rate'], ddof=1))
```

# Standard Deviations

```python
print(np.std(df['age'], ddof=1))
print(np.std(df['salary'], ddof=1))
print(np.std(df['happiness_rate'], ddof=1))
```
