````python
features = pd.read_csv('features')

targets = pd.read_csv('target')

features
````

````python
targets
````

````python
first_merged_data = pd.merge(features, targets, left_on= 'id', right_on ='id')
first_merged_data
````

````python
second_merged_data = pd.merge(targets, features, left_on= 'id', right_on ='id')
second_merged_data
````

````python
outer_merged_data = pd.merge(features, targets, left_on ='id', right_on = 'id', how ='outer')
outer_merged_data
````

````python
left_merged_data = pd.merge(features, targets, left_on ='id', right_on = 'id', how ='left')
left_merged_data
````

````python
right_merged_data = pd.merge(features, targets, left_on ='id', right_on = 'id', how ='right')
right_merged_data
````

````python
print(len(first_merged_data))
print(len(second_merged_data))
print(len(outer_merged_data))
print(len(left_merged_data))
print(len(right_merged_data))
````
