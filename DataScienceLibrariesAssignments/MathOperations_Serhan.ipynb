```python
for i in range (0,5):
  print("Mean of the " + str(i+1) + "th row is : "
   + str(np.mean(arr[:, i])))
```

```python
for i in range (0,5):
  print("Std deviation of the " + str(i+1) + "th row is : "
   + str(np.std(arr[:, i])))
```
