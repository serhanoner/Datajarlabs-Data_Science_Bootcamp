```python
import pandas as pd
import numpy as np
```

```python
from pandas import read_csv
data = read_csv('train.csv')
print(data)
```

```python
s = pd.DataFrame(data)
s
```

```python
s.groupby("Sex")["Age"].count()
```

```python
girls = s.loc[s["Sex"] == 'female', 'Age'].sum()
```

```python
bois = s.loc[s["Sex"] == 'male', 'Age'].sum()
```

```python
print("Avg age of girls is : " , girls/261)
```

```python
print("Avg age of bois is : " , bois/453)
```

```python
# now we're dealing with passenger class
s.groupby("Pclass")["Age"].count()
```

```python
first_class = s.loc[s["Pclass"] == 1 , 'Age'].sum()
first_class
```

```python
second_class = s.loc[s["Pclass"] == 2, 'Age'].sum()
second_class
```

```python
third_class = s.loc[s["Pclass"] == 3 , 'Age'].sum()
third_class
```

```python
print("Avg age of first class is : " , first_class/186)
print("Avg age of second class is : " , second_class/173)
print("Avg age of third class is : " , third_class/355)
```
