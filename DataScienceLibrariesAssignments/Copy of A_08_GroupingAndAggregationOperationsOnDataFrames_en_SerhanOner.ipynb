**(1)** Calculate the average age of the passengers for each gender and passenger class by using `groupby()` method.

```python
import pandas as pd
import numpy as np

from pandas import read_csv
data = read_csv('train.csv')

df = pd.DataFrame(data)

df.groupby("Sex")["Age"].mean()
```
