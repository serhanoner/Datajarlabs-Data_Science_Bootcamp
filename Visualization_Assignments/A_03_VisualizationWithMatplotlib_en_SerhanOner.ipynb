In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
spain = df[df['location'] == 'Spain'].tail(100)
france = df[df['location'] == 'France'].tail(100)
germany = df[df['location'] == 'Germany'].tail(100)
italy = df[df['location'] == 'Italy'].tail(100)
spain
```

```python
import numpy as np

plt.figure(figsize=(25,10))

plt.suptitle('Total Cases and Total Cases for 4 Countries',
             fontstyle = 'normal',
             fontsize = 20,
             color = "darkblue")

x_ax = np.arange(0,100)

spain_cases = df.loc[df['location'] == 'Spain', 'total_cases'].tail(100)
france_cases = df.loc[df['location'] == 'France', 'total_cases'].tail(100)
germany_cases = df.loc[df['location'] == 'Germany', 'total_cases'].tail(100)
italy_cases = df.loc[df['location'] == 'Italy', 'total_cases'].tail(100)

spain_deaths = df.loc[df['location'] == 'Spain', 'total_deaths'].tail(100)
france_deaths = df.loc[df['location'] == 'France', 'total_deaths'].tail(100)
germany_deaths = df.loc[df['location'] == 'Germany', 'total_deaths'].tail(100)
italy_deaths = df.loc[df['location'] == 'Italy', 'total_deaths'].tail(100)



plt.subplot(1,2,1)

plt.title('Total Cases for 4 Countries',
          fontstyle = 'normal',
          fontsize = 15,
          color = "darkblue")
plt.xlabel('Last 100 Days',
           fontstyle = 'normal',
           fontsize = 10,
           color = 'darkred')
plt.ylabel('Total Cases',
           fontstyle = 'normal',
           fontsize = 10,
           color = 'darkred')
plt.xticks(rotation = 45, fontsize = 10)

plt.plot(x_ax, spain_cases , c = "blue" , linewidth=3, linestyle="-" )
plt.plot(x_ax, france_cases , c = "cyan" , linewidth=3, linestyle="--" )
plt.plot(x_ax, germany_cases , c = "green", linewidth=3, linestyle=":" )
plt.plot(x_ax, italy_cases , c = "red", linewidth=3, linestyle="-." )

plt.legend(['Spain', 'France', 'Germany', 'Italy'], title = 'Countries', loc = 4)


plt.subplot(1,2,2)

plt.title('Total Deaths for 4 Countries',
          fontstyle = 'normal',
          fontsize = 15,
          color = "darkblue")
plt.xlabel('Last 100 Days',
           fontstyle = 'normal',
           fontsize = 10,
           color = 'darkred')
plt.ylabel('Total Deaths',
           fontstyle = 'normal',
           fontsize = 10,
           color = 'darkred')

plt.xticks(rotation = 45, fontsize = 10)

plt.plot(x_ax, spain_deaths , c = "blue" , linewidth=3, linestyle="-" )
plt.plot(x_ax, france_deaths , c = "cyan" , linewidth=3, linestyle="--" )
plt.plot(x_ax, germany_deaths , c = "green", linewidth=3, linestyle=":" )
plt.plot(x_ax, italy_deaths , c = "red", linewidth=3, linestyle="-." )


plt.legend(['Spain', 'France', 'Germany', 'Italy'], title = 'Countries', loc = 4)


plt.show()
```
