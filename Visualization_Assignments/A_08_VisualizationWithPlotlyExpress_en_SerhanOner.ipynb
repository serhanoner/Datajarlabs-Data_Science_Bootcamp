In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Plotly Express. 

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
import plotly.express as px
```

**(1)** Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
df2 = df[df['date'] >= '2020-04-01']
df2.head()
```

```python
data = df2[df2.location.isin(['Spain', 'France', 'Germany', 'Italy'])]
fig = px.line(data, x="date", y="total_deaths", title='Total Deaths by Date in 4 Countries', color = 'location')
fig.show()
```

```python
data = df2[df2.location == 'Italy']
fig = px.line(data, x="date", y="total_deaths", title='Total Deaths by Date in 4 Countries')
fig.show()
```

**(2)** Plot a bar plot with plotly express for average death number that compares continents.

```python
avg_deaths = df2.groupby(['continent', 'date']).new_deaths.mean().reset_index()
avg_deaths
```

```python
avg_deaths2 = df2.groupby(['continent', 'date']).new_deaths.sum().reset_index()
avg_deaths2
```

```python
fig = px.bar(data_frame=avg_deaths, 
             x='date', y='new_deaths', color = 'continent')

fig.show()
```

```python
#below is the grouped version, but it looks like there's nothing to _observe_ here.

fig = px.bar(data_frame=avg_deaths, 
             x='date', y='new_deaths', color = 'continent',
             barmode = 'group')
fig.show()
```

**(3)** Plot a histogram for daily deaths for any country you choose. 

```python
avg_deaths2 = df2.groupby(['location', 'date']).new_deaths.sum().reset_index()
avg_deaths2
```

```python
df3 = avg_deaths2[avg_deaths2['location'] == 'Turkey']
df3
```

```python
fig = px.histogram(df3, x="date", color='new_deaths', marginal='box')
fig.show()
```
