
**(1)** Create a blank figure with matplotlib functions. Plot cosine function for the numbers beetwen -3$\pi$ and 3$\pi$ with 0.1 intervals.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

fig = plt.figure()
plt.title('An empty figure')
```

```python
x = np.arange(math.pi * (-3), math.pi * 3 , 0.1)
y = np.cos(x)

fig = plt.figure()

ax = fig.subplots()

ax.plot(x,y)
plt.show()
```

**(2)** This time, create the same plot in red color with object methods. 

```python
x = np.arange(math.pi * (-3), math.pi * 3 , 0.1)
y = np.cos(x)

fig = plt.figure()

ax = fig.subplots()

ax.plot(x,y, color = '#EE0000')
plt.show()
```
